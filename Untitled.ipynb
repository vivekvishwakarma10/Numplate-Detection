```python
import os
import pandas as pd
import numpy as np
import plotly.express as px
from skimage import io
# from paddleocr import PaddleOCR
import cv2
import xml.etree.ElementTree as xet
from glob import glob
```

```python
path = glob('Text/*.xml')
labels_dict = dict(filepath=[],xmin=[],xmax=[],ymin=[],ymax=[])
for filename in path:

    info = xet.parse(filename)
    root = info.getroot()
    member_object = root.find('object')
    labels_info = member_object.find('bndbox')
    xmin = float(labels_info.find('xmin').text)
    xmax = float(labels_info.find('xmax').text)
    ymin = float(labels_info.find('ymin').text)
    ymax = float(labels_info.find('ymax').text)

    labels_dict['filepath'].append(filename)
    labels_dict['xmin'].append(xmin)
    labels_dict['xmax'].append(xmax)
    labels_dict['ymin'].append(ymin)
    labels_dict['ymax'].append(ymax)
```

```python
df = pd.DataFrame(labels_dict)
df.to_csv('labels.csv',index=False)
df.head(15)
```

```python
filename = df['filepath'][0]
def getFilename(filename):
    filename_image = xet.parse(filename).getroot().find('filename').text
    filepath_image = os.path.join('../First Project/Text',filename_image)
    return filepath_image
getFilename(filename)
```
