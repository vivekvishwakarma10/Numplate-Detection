```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
import easyocr
```

```python
path = './ANPRwithPython/increased resolu (1).png'
img01 = cv2.imread(path)
gray = cv2.cvtColor(img01, cv2.COLOR_BGR2GRAY)
th = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
kernel = np.ones((1, 1), np.uint8)
th = cv2.erode(th, kernel, iterations=1)
th = cv2.dilate(th, kernel, iterations=1)
th = cv2.medianBlur(th, 3)
```

```python
plt.imshow(th)
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(th)
result
```

```python
image = cv2.imread(path)
```

```python
def preprocess_image(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply Gaussian blur to remove noise
    blur = cv2.GaussianBlur(gray, (1, 1), 100)
    
    # Apply adaptive thresholding to enhance contrast
    thresh = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 33, 2)
    
    # Apply dilation to make characters thicker
    kernel = np.ones((2, 2), np.uint8)
    dilate = cv2.dilate(thresh, kernel, iterations=1)
    
    # Apply erosion to remove small gaps between characters
    erode = cv2.erode(dilate, kernel, iterations=1)
    
    return erode
```

```python
preprocess_image(image)
```

```python
plt.imshow(preprocess_image(image))
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(preprocess_image(image))
result
```
